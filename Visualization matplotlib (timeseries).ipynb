```python
import pandas as pd
import numpy as np
```

```python
tiempo = pd.read_csv('https://goo.gl/ccoFyG', skiprows=1)
```

```python
tiempo.columns = ['fecha', 'dieta', 'gimnasio', 'finanzas']
```

```python
tiempo.head()
```

```python
tiempo.shape
```

```python
tiempo.dtypes
```

```python
tiempo['fecha'] = pd.to_datetime(tiempo['fecha'])
```

```python
tiempo.dtypes
```

```python
tiempo.head()
```

```python
tiempo.set_index('fecha', inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tiempo[['dieta']].plot(figsize=(16,8), color='royalblue', linewidth=2, fontsize=14)
plt.xlabel('AÃ±os', fontsize=14)
```

```python
tiempo[['gimnasio']].plot(figsize=(16,8), color='darkorange', linewidth=2, fontsize=14)
plt.xlabel('AÃ±os', fontsize=14)
```

```python
tiempo[['finanzas']].plot(figsize=(16,8), color="green", linewidth=2, fontsize=14)
plt.xlabel('AÃ±os', fontsize=14)
```

```python
tiempo["dieta"].rolling(4).mean().plot(figsize=(16,8), color='royalblue', linewidth=2, fontsize=14)
plt.xlabel('AÃ±os', fontsize=14)
```

```python
tiempo["finanzas"].rolling(8).mean().plot(figsize=(16,8), color='g', linewidth=2, fontsize=14)
plt.xlabel('AÃ±os', fontsize=14) # EL rolling estÃ¡ pensado para buscar una endencia
```

```python
tiempo["dieta"].diff().plot(figsize=(16,8), color='royalblue',linewidth=2, fontsize=14)
plt.xlabel('AÃ±os', fontsize=14)
```

```python
tiempo["gimnasio"].diff().plot(figsize=(16,8), color='darkorange', linewidth=2, fontsize=14)
plt.xlabel('AÃ±os', fontsize=14)
```

```python
tiempo.diff().plot(figsize=(16,8), linewidth=2, fontsize=14)
plt.xlabel('AÃ±os)', fontsize=14)
```
