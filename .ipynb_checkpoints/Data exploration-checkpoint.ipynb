```python
df= pd.read_csv("http://goo.gl/rxM4GY")
df.head()
```

```python
df.shape
```

```python
dfdeportes= df[df.Genre=="Sports"]
dfdeportes
```

```python
dfdep4= dfdeportes[dfdeportes.EU_Sales>4]
dfdep4
```

```python
dfdep4.sort_values("Name")
```

```python
contWii=0;
contPs4=0;
contPs3=0;
for row in dfdep4["Platform"]:
    if(row=="Wii"):
        contWii+=1;
    elif(row=="PS4"):
        contPs4+=1;
    else:
        contPs3+=1;
print("Wii: " +  str(contWii));
print("Ps4: " + str(contPs4));
print("contPs3: " + str(contPs3));
```
